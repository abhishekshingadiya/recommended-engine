```
import gzip
def parse(path):
  g = gzip.open(path, 'rb')
  for l in g:
    yield eval(l)
def getDF(path):
  i = 0
  df = {}
  for d in parse(path):
    df[i] = d
    i += 1
  return pd.DataFrame.from_dict(df, orient='index')
```

```
import requests 
file_url = "http://snap.stanford.edu/data/amazon/productGraph/categoryFiles/reviews_Baby_5.json.gz"

r = requests.get(file_url, stream = True) 

with open("reviews_Baby_5.json.gz","wb") as pdf: 
	for chunk in r.iter_content(chunk_size=1024):
		# writing one chunk at a time to pdf file 
		if chunk: 
			pdf.write(chunk)
```

```
import pandas as pd 
df1= getDF('reviews_Baby_5.json.gz')
df1=df1[["reviewerID","asin","reviewText","overall"]]
df1.rename(columns={'reviewerID':'userId','asin':'prodId','overall':'rating'}, inplace=True)
df2= getDF('reviews_Beauty_5.json.gz')
df2=df2[["reviewerID","asin","reviewText","overall"]]
df2.rename(columns={'reviewerID':'userId','asin':'prodId','overall':'rating'}, inplace=True)
df3= getDF('reviews_Automotive_5.json.gz')
df3=df3[["reviewerID","asin","reviewText","overall"]]
df3.rename(columns={'reviewerID':'userId','asin':'prodId','overall':'rating'}, inplace=True)
df4= getDF('reviews_Health_and_Personal_Care_5.json.gz')
df4=df4[["reviewerID","asin","reviewText","overall"]]
df4.rename(columns={'reviewerID':'userId','asin':'prodId','overall':'rating'}, inplace=True)
df5= getDF('reviews_Pet_Supplies_5.json.gz')
df5=df5[["reviewerID","asin","reviewText","overall"]]
df5.rename(columns={'reviewerID':'userId','asin':'prodId','overall':'rating'}, inplace=True)
# df1= pd.read_csv('reviews_Baby.csv',names=["userId", "prodId", "rating", "timestamp"])
# df2= pd.read_csv('ratings_Beauty.csv',names=["userId", "prodId", "rating", "timestamp"])
# df3= pd.read_csv('ratings_Automotive.csv',names=["userId", "prodId", "rating", "timestamp"])
# df4= pd.read_csv('ratings_Health_and_Personal_Care.csv',names=["userId", "prodId", "rating", "timestamp"])
# df5= pd.read_csv('ratings_Pet_Supplies.csv',names=["userId", "prodId", "rating", "timestamp"])
```

```
idx1 = pd.Index(df1.userId)
idx2 = pd.Index(df2.userId)
idx3 = pd.Index(df3.userId)
idx4 = pd.Index(df4.userId)
idx5 = pd.Index(df5.userId)
idx12=idx2.intersection(idx1)
idx34=idx4.intersection(idx3)
idx125=idx12.intersection(idx5)
finalidx=idx125.intersection(idx34)
```
